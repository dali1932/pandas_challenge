PyCitySchool Analysis - Based on School Performance

```python
import pandas as pd
from pathlib import Path

school_file_to_load = Path("Resources/schools_complete.csv")
student_file_to_load = Path("Resources/students_complete.csv")

school_file = pd.read_csv(school_file_to_load)
student_file = pd.read_csv(student_file_to_load)

data_merged = pd.merge(school_file, student_file, how ="right", on =["school_name", "school_name"])
data_merged.head()

```

```python
school_unique = data_merged["school_name"].unique()
school_count = len(school_unique)
school_count

# school_count = len(set(data_merged["school_name"]))--->using set(elements), duplicates are automatically removed
# school_count
```

```python
student_unique = data_merged["Student ID"].unique()
student_count = len(student_unique)
student_count
```

```python
data_school = data_merged[["School ID","school_name","budget"]]
unique_school=data_school.drop_duplicates(["School ID"]).reset_index()
# unique_school
total_budget=unique_school["budget"].sum()
total_budget
```

```python
avg_math_score = data_merged["math_score"].mean()
avg_math_score
```

```python
avg_reading_score = data_merged["reading_score"].mean()
avg_reading_score
```

```python
# Use the following to calculate the percentage of students who passed math (math scores greather than or equal to 70)
passing_math_count = data_merged[(data_merged["math_score"] >= 70)].shape[0] 
#shape[0] gives you the number of rows or observations in the DataFrame
passing_math_percentage = passing_math_count / float(student_count)*100
passing_math_percentage
```

```python
passing_reading_count = data_merged[(data_merged["reading_score"] >= 70)].shape[0] 
#shape[0] gives you the number of rows or observations in the DataFrame
passing_reading_percentage = passing_reading_count / float(student_count)*100
passing_reading_percentage
```

```python
overall_passing_count = data_merged[(data_merged["reading_score"] >= 70) & (data_merged["math_score"] >= 70)].shape[0]
overall_passing_percentage = overall_passing_count / float(student_count) *100
overall_passing_percentage
```

```python
# Create a high-level snapshot of the district's key metrics in a DataFrame
district_summary = pd.DataFrame({"Total School":[school_count],"Total Students":[student_count], 
                                 "Total Budget":[total_budget],"Average Math Score":[avg_math_score],
                                 "Average Reading Score":[avg_reading_score],"% Passing Math":[passing_math_percentage],
                            	"% Passing Reading":[passing_reading_percentage],"% Overall Passing":[overall_passing_percentage]})
# Formatting
district_summary["Total Students"] = district_summary["Total Students"].map("{:,}".format)
district_summary["Total Budget"] = district_summary["Total Budget"].map("${:,}".format)
district_summary["Average Math Score"] = district_summary["Average Math Score"].map("{:,.2F}".format)
district_summary["Average Reading Score"] = district_summary["Average Reading Score"].map("{:,.2F}".format)
district_summary["% Passing Math"] = district_summary["% Passing Math"].map("{:,.2F}%".format)
district_summary["% Passing Reading"] = district_summary["% Passing Reading"].map("{:,.2F}%".format)
district_summary["% Overall Passing"] = district_summary["% Overall Passing"].map("{:,.2F}%".format)

# Display the DataFrame
district_summary
```

```python
# Calculate the average test scores per school from school_data_complete
per_school_math = data_merged.groupby('school_name')['math_score'].mean()

per_school_reading = data_merged.groupby('school_name')['reading_score'].mean()
score=pd.DataFrame({'Avg math':per_school_math.round(2), 'Avg reading':per_school_reading.round(2)})
score
```

```python
# Calculate the number of students per school with math scores of 70 or higher from school_data_complete
school_students_passing_math = data_merged[data_merged["math_score"] >= 70].groupby('school_name').size()

school_students_passing_math 
```

```python
# Calculate the number of students per school with reading scores of 70 or higher from school_data_complete
school_students_passing_reading = data_merged[data_merged["reading_score"] >= 70].groupby('school_name').size()

school_students_passing_reading 
```

```python
# Use the provided code to calculate the number of students per school that passed both math and reading with scores of 70 or higher
students_passing_math_and_reading = data_merged[(data_merged["math_score"] >= 70)&(data_merged["reading_score"] >= 70)]
school_students_passing_math_and_reading = students_passing_math_and_reading.groupby(["school_name"]).size()
school_students_passing_math_and_reading
```

```python
# Use the provided code to calculate the passing rates
per_school_passing_math = school_students_passing_math / student_per_school * 100
per_school_passing_reading = school_students_passing_reading / student_per_school * 100
overall_passing_rate = school_students_passing_math_and_reading / student_per_school * 100
passing=pd.DataFrame({'% pass math':per_school_passing_math.map('{:.2f}%'.format), '% pass reading':per_school_passing_reading.map('{:.2f}%'.format),
                      '% pass both':overall_passing_rate.map('{:.2f}%'.format)})
passing
```

```python
# Create a DataFrame called `per_school_summary` with columns for the calculations above.
per_school_summary = pd.DataFrame({'Type':school_types,'Student Count':student_per_school,
                                   'Total School Budget':per_school_budget,
                                   'Per Student Budget':per_school_capita,'Avg math':per_school_math.round(2), 'Avg reading':per_school_reading.round(2),
                                   'student count passing math':school_students_passing_math,
                                   'student count passing reading':school_students_passing_reading,  
    'student count passing both':school_students_passing_math_and_reading,
                                   '% pass math':per_school_passing_math, '% pass reading':per_school_passing_reading,
                      '% pass both':overall_passing_rate})

# Formatting
per_school_summary["Total School Budget"] = per_school_summary["Total School Budget"].map("${:,.2f}".format)
per_school_summary["Per Student Budget"] = per_school_summary["Per Student Budget"]

# Display the DataFrame
per_school_summary
```

Scores by School Spending

```python
# Establish the bins
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
# Create a copy of the school summary since it has the "Per Student Budget"
school_spending_df = per_school_summary.copy()
# school_spending_df["Per Student Budget"] = pd.to_numeric(school_spending_df["Per Student Budget"])
```

```python
# Use `pd.cut` to categorize spending based on the bins.
school_spending_df["Per Student Budget"] =pd.cut(school_spending_df["Per Student Budget"],bins=spending_bins,
                                                 labels=labels,include_lowest=True)
school_spending_df
```

```python
#  Calculate averages for the desired columns.
spending_math_scores = school_spending_df.groupby(["Per Student Budget"], observed=True)["Avg math"].mean()
spending_reading_scores = school_spending_df.groupby(["Per Student Budget"], observed=True)["Avg reading"].mean()
spending_passing_math = school_spending_df.groupby(["Per Student Budget"], observed=True)["% pass math"].mean()
spending_passing_reading = school_spending_df.groupby(["Per Student Budget"], observed=True)["% pass reading"].mean()
overall_passing_spending = school_spending_df.groupby(["Per Student Budget"], observed=True)["% pass both"].mean()
```

```python
# Assemble into DataFrame
spending_summary =pd.DataFrame({'spending_math_scores':spending_math_scores,'spending_reading_scores':spending_reading_scores,
                                'spending_passing_math':spending_passing_math,'spending_passing_reading':spending_passing_reading,
                                'overall_passing_spending':overall_passing_spending})

# Display results
spending_summary
```
